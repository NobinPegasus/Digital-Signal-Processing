

**$X_k = \sum_{n=0}^{N-1}{x_n\cdot e^{-i2\pi{kn/N}}} = \sum_{n=0}^{N-1}{x_n[cos(2\pi{kn/N}) -i\cdot sin(2\pi{kn/N})]}$**






 

```python
def DFT(x):


    N = len(x)
    # print(x)
    n = np.arange(N)
    # print(type(n))
    # print(n)
    k = n.reshape((N, 1))
    # print(type(k))
    # print(k)
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)
    
    return X
```

```python
X = DFT(x)


N = len(X)
# print(N)
n = np.arange(N)
# print(n)
T = N/s_r
# print(T)
freq = n/T 
# print(type(freq))

# [freq for i in freq if(i<(s_r//2))]


# plt.stem(X)
freq=np.array([i for i in freq if(i<(s_r//2))])
# print(freq)
X=X[:len(freq)]
print(X)
print(freq)

nn = len(freq)
div = freq[nn-1]
X=X[:]/div

plt.figure(figsize = (8, 6))
# plt.plot(freq, np.abs(X), 'r')
# print(X[1])
# print(np.abs(X[1]))
plt.stem(freq, np.abs(X), 'r', markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.xlim(0, 11)
plt.xticks(np.arange(0,11))
plt.grid(True)
plt.show()
```
